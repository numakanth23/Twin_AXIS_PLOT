```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import xarray as xr
w=pd.read_csv("WYO_JUL22_2019.csv")
w1 = w[0:72]
plt.plot(w1['TEMP(C)'],(w1['HGHT(m)']/1000))
plt.xlabel('Temp(c)')
plt.ylabel('Height(KM)')
plt.savefig('JUL22_2019_TempVSHGHT.png')
#plt.xlim(-50,50)
w1
```

```python
w1.columns
plt.plot(w1['RELH(%)'],(w1['HGHT(m)']/1000))
plt.xlabel('RELH(%)')
plt.ylabel('Height(KM)')
plt.savefig('JUL22_2019_RHVSHGHT.png')
```

```python
import matplotlib.pyplot as plt

fig=plt.figure(figsize = (12,8))
ax1=fig.add_subplot(111, label="1", frame_on=True)
ax2=fig.add_subplot(111, label="2", frame_on=False)
# ax3=fig.add_subplot(111, label="3", frame_on=False)

ax1.plot(w1['TEMP(C)'],(w1['HGHT(m)']/1000), color="red")
ax1.set_xlabel("Temperature(C)", color="red")
ax1.set_ylabel("Altitude(KM)", color="red")
ax1.tick_params(axis='x', colors="red")
ax1.tick_params(axis='y', colors="red")
ax2.plot(w1['RELH(%)'],(w1['PRES(hpa)']), color="blue")
ax2.xaxis.tick_top()
ax2.yaxis.tick_right()
ax2.set_xlabel('Relative Humidity(%)', color="blue") 
ax2.set_ylabel('Pressure (hPa)', color="blue")       
ax2.xaxis.set_label_position('top') 
ax2.yaxis.set_label_position('right') 
ax2.tick_params(axis='x', colors="blue")
ax2.tick_params(axis='y', colors="blue")

ax1.spines['bottom'].set_color("red")
ax1.spines['top'].set_color('blue')
ax1.spines['right'].set_color('blue')
ax1.spines['left'].set_color("red")
plt.savefig('JUL22_2019_TempVSHGHT_RH_Pres.png')
plt.show()
```
